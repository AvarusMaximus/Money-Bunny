```python
from not_bunny import *
import pandas as pd
```

```python
data = pd.read_csv('stock_data.csv')
del data['Volume']
data = data.set_index(['Date', 'Ticker'])
data
```

```python
ticker_list = data.index.levels[1].tolist()
ticker_list
```

```python
# Creating an empty dataframe that contains ticker and mae values of each model (4 columns for 4 models) for each company in a row
mae_df = pd.DataFrame(columns=['Ticker', 'ANN', 'LSTM_1D', 'LSTM_2D', 'LSTM_3D'])
r2_df = pd.DataFrame(columns=['Ticker', 'ANN', 'LSTM_1D', 'LSTM_2D', 'LSTM_3D'])
mse_df = pd.DataFrame(columns=['Ticker', 'ANN', 'LSTM_1D', 'LSTM_2D', 'LSTM_3D'])

for ticker in ticker_list:
    company = data.xs(ticker, level=1)
    adj_close = company['Adj Close'].values.reshape(-1, 1)
    mae_ann, mse_ann, r2_ann, pred_ann, actual_ann = ann(adj_close) # Mean Absolute Error of ANN
    mae_lstm_1d, mse_lstm_1d, r2_lstm_1d, pred_lstm_1d, actual_lstm_1d = lstm_1d(adj_close) # Mean Absolute Error of LSTM_1D
    mae_lstm_2d, mse_lstm_2d, r2_lstm_2d, pred_lstm_2d, actual_lstm_2d = lstm_2d(adj_close) # Mean Absolute Error of LSTM_2D
    mae_lstm_3d, mse_lstm_3d, r2_lstm_3d, pred_lstm_3d, actual_lstm_3d = lstm_3d(adj_close) # Mean Absolute Error of LSTM_3D
    mae_df = pd.concat([mae_df, pd.DataFrame({'Ticker': ticker, 'ANN': mae_ann, 'LSTM_1D': mae_lstm_1d, 'LSTM_2D': mae_lstm_2d, 'LSTM_3D': mae_lstm_3d}, index = [0])], axis=0)
    r2_df = pd.concat([r2_df, pd.DataFrame({'Ticker': ticker, 'ANN': r2_ann, 'LSTM_1D': r2_lstm_1d, 'LSTM_2D': r2_lstm_2d, 'LSTM_3D': r2_lstm_3d}, index = [0])], axis=0)
    mse_df = pd.concat([mse_df, pd.DataFrame({'Ticker': ticker, 'ANN': mse_ann, 'LSTM_1D': mse_lstm_1d, 'LSTM_2D': mse_lstm_2d, 'LSTM_3D': mse_lstm_3d}, index = [0])], axis=0)
```
