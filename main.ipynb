```python
from not_bunny import *
import pandas as pd
import numpy as np
from sklearn.preprocessing import MinMaxScaler
```

```python
data = pd.read_csv('stock_data.csv')
del data['Volume']
data = data.set_index(['Date', 'Ticker'])
data
```

```python
ticker_list = data.index.levels[1].tolist()
ticker_list
```

```python
# Creating an empty dataframe that contains ticker and mae values of each model (4 columns for 4 models) for each company in a row
mae_df = pd.DataFrame(columns=['Ticker', 'ANN', 'LSTM_1D', 'LSTM_2D', 'LSTM_3D'])

for ticker in ticker_list:
    company = data.xs(ticker, level=1)
    # Scaling the Adj Close values
    scaler = MinMaxScaler(feature_range=(0, 1))
    scaled_price = scaler.fit_transform(company['Adj Close'].values.reshape(-1, 1))
    mae_ann = ann(scaled_price) # Mean Absolute Error of ANN
    mae_lstm_1d = lstm_1d(scaled_price) # Mean Absolute Error of LSTM_1D
    mae_lstm_2d = lstm_2d(scaled_price) # Mean Absolute Error of LSTM_2D
    mae_lstm_3d = lstm_3d(scaled_price) # Mean Absolute Error of LSTM_3D
    mae_df = pd.concat([mae_df, pd.DataFrame({'Ticker': ticker, 'ANN': mae_ann, 'LSTM_1D': mae_lstm_1d, 'LSTM_2D': mae_lstm_2d, 'LSTM_3D': mae_lstm_3d})], ignore_index=True)
```

```python
mae_df.to_csv('mae.csv', index=False) # Exporting the model performance, in the form of minimum absolute error for each company to a csv file
```
