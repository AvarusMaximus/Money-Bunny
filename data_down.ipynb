```python
import pandas as pd 
import yfinance as yf
```

```python
tickers = ['MSFT', 'AAPL', 'GOOGL', 'AMZN', 'META', 'TSLA', 'NVDA', 'AMD', 'CRM', 'INTC']
end_date = pd.to_datetime('2024-02-02')
start_date = end_date - pd.DateOffset(years=8)

data = yf.download(tickers, start=start_date, end=end_date).stack()
data
```

```python
data.index.names = ['Date', 'Ticker']
data
```

```python
data.to_csv('stock_data.csv')
```
